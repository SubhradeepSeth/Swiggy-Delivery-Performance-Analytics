```python
avg_age = df['Delivery_person_Age'].mean().round()
```

```python
print(avg_age)
```

```python
df['Delivery_person_Age']=df['Delivery_person_Age'].fillna(avg_age)
```

```python
avg_rating = df['Delivery_person_Ratings'].mean().round(1)
```

```python
print(avg_rating)
```

```python
df['Delivery_person_Ratings']=df['Delivery_person_Ratings'].fillna(avg_rating)
```

```python
df['Time_Orderd'] = df['Time_Orderd'].ffill()
```

```python
df['Weather_conditions'].unique()
```

```python
most_occ_weather = df['Weather_conditions'].mode()[0]
```

```python
print(most_occ_weather)
```

```python
df['Weather_conditions']= df['Weather_conditions'].fillna('most_occ_weather')
```

```python
df['Road_traffic_density'].unique()
```

```python
df['Weather_conditions']= df['Weather_conditions'].replace('most_occ_weather',most_occ_weather)
```

```python
most_occ_traffic = df['Road_traffic_density'].mode()[0]
```

```python
print(most_occ_traffic)
```

```python
df['Road_traffic_density']=df['Road_traffic_density'].fillna('most_occ_traffic')
```

```python
df['Road_traffic_density']=df['Road_traffic_density'].replace('most_occ_traffic',most_occ_traffic)
```

```python
df['multiple_deliveries'].unique()
```

```python
avg_mul_del=df['multiple_deliveries'].mean().round()
```

```python
print(avg_mul_del)
```

```python
df['multiple_deliveries']=df['multiple_deliveries'].fillna(avg_mul_del)
```
